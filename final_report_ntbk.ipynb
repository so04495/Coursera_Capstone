```python
import requests
import json
import numpy as np
from pandas.io.json import json_normalize
import folium as f
```

```python
url = 'https://cocl.us/new_york_dataset'
nyd = requests.get(url).json()
```

```python
df = json_normalize(nyd['features'])
del df['id']
del df['type']
del df['geometry_name']
df = df[df['geometry.type'] == 'Point']
del df['properties.annoline1']
del df['properties.annoline2']
del df['properties.annoline3']
del df['properties.annoangle']
del df['properties.bbox']
df['geometry.coordinates'] = df['geometry.coordinates'].astype(str)
df[['Lat', 'Lon']] = df['geometry.coordinates'].str.strip(r'[]').str.split(',', expand=True).astype(np.float)
del df['geometry.coordinates']
df
```

```python
map1 = f.Map(location=[40.8, -74], tiles='Stamen Terrain')
for i, e in df.iterrows():
    f.CircleMarker(
        location=[e['Lon'], e['Lat']],
        radius=1,
        color="green"
    ).add_to(map1)
map1
```
